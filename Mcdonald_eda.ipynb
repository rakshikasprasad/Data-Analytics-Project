Data normalization:

```python
# Adding a new columns based on normalization per serve grams or ml will give us more objective findings
# Consequently,the multiplication of these values to 100 gives us more vissulaly objective values 
# Assumption of the 1ml=1g was made to simplify the the normalization


for col in df['Energy (kCal)']:
    df['Energy normalized'] = (df['Energy (kCal)'] / df['Per Serve Size'])*100
    
for col in df['Protein (g)']:
    df['Protein normalized'] = (df['Protein (g)'] / df['Per Serve Size'])*100 

for col in df['Total fat (g)']:
    df['Total fat normalized'] = (df['Total fat (g)'] / df['Per Serve Size'])*100
    
# Note that Cholesterols in mg, so thus, it will be multiplied to 0.1 g
    
for col in df['Cholesterols (mg)']:
    df['Cholesterols normalized'] = (df['Cholesterols (mg)'] / df['Per Serve Size'])*0.1
    
for col in df['Total carbohydrate (g)']:
    df['Carbohydrate normalized'] = (df['Total carbohydrate (g)'] / df['Per Serve Size'])*100  
    
for col in df['Total carbohydrate (g)']:
    df['Carbohydrate normalized'] = (df['Total carbohydrate (g)'] / df['Per Serve Size'])*100
    
for col in df['Total Sugars (g)']:
    df['Sugar normalized'] = (df['Total Sugars (g)'] / df['Per Serve Size'])*100
    
# Note that Sodium in mg, so thus, it will be multiplied to 0.1 g

for col in df['Sodium (mg)']:
    df['Sodium normalized'] = (df['Sodium (mg)']/ df['Per Serve Size'])*0.1
```

We have used data normalization.

Calories & Sugar Level Per Serve:

```python
cols = ["Energy (kCal)","Protein (g)","Total fat (g)","Total carbohydrate (g)","Total Sugars (g)","Sodium (mg)"]

df["Per Serve Size_No units"] = df["Per Serve Size"].map(lambda x: str(x)[:-2])

for col in cols:
    df[col+"mean"] = df[col].astype(float) / df["Per Serve Size_No units"].astype(float)

df.head()
```
